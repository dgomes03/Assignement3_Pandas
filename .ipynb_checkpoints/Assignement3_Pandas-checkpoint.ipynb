### Read the file:

```python
loans = pd.read_csv('/Users/diogogomes/Documents/Uni/Introdução ao Python/Assignement3_Pandas/data/kiva_loans.csv')
loans
```

### 1. Select rows using string methods (e.g. .isin) and numerical conditions

Filtering the dataframe to show only the country Nigeria and a loan_amount grater than 3000:

```python
filtered_loans = loans[(loans['country'].isin(['Nigeria'])) & (loans['loan_amount'] > 3000)]
filtered_loans
```

### 2. Sort table using one or more columns

Sort the table by loan_amount in a descending order:

```python
sorted_loans = loans.sort_values(by='loan_amount', ascending=False)
sorted_loans
```

Sort by activity and term_in_months. This will return the activity in alphabetical order and the term_in_months in descending order.

```python
sorted_loans2 = loans.sort_values(by=['activity', 'term_in_months'], ascending=[True, False])
sorted_loans2
```

### 3. Group and aggregate rows with "groupby"

Group the table by country, aggregating the values loan_amount into a mean value, and also counting the number of loans each country has:

```python
loans_group = loans.groupby('country').agg(
    avg_loan_amount=('loan_amount', 'mean'),
    loan_count=('loan_amount', 'size')
).reset_index()

loans_group
```

### 4. Create a new variable with "apply" and "lambda"

Let's create a new variable that categorizes the loans into three categories. 'small loan' if 'loan_amount' is lower than 5000, 'medium loan' if it'slower than 30000 and 'big loan' if it's higher than 30000.

```python
loans['loan_amount_category'] = loans.apply(
    lambda row: (
        'small loan' if row['loan_amount'] < 5000 else
        'medium loan' if row['loan_amount'] < 30000 else
        'big loan'
    ),
    axis=1
)
loans
```

Now let's sort the loan_amount to see if the categories worked:

```python
loan_category_sorted = loans.sort_values(by='loan_amount', ascending = False)
loan_category_sorted
```
